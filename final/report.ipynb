# Artificial Life Evolution Plots

```python
import matplotlib.pyplot as plt
import pickle
```

```python
with open('epoch_results.pkl', 'rb') as file:
    epoch_results = pickle.load(file)

# Extract the max and average mobility data
max_mobility = [result['max_height'] for result in epoch_results]
average_mobility = [result['average_mobility'] for result in epoch_results]
epochs = [result['epoch'] for result in epoch_results]

# Plot the data
plt.figure(figsize=(14, 7))
plt.plot(epochs, max_mobility, label='Max Mobility', marker='o', linestyle='-', color='red')
plt.plot(epochs, average_mobility, label='Average Mobility', marker='s', linestyle='--', color='blue')
plt.title('Max and Average Mobility over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Mobility')
plt.legend()
plt.grid(True)
plt.show()
```
